### 5) Extract {𝑌𝑒𝑎𝑟, 𝑀𝑜𝑛𝑡ℎ, 𝑊𝑒𝑒𝑘, 𝑊𝑒𝑒𝑘𝐷𝑎𝑦, 𝐷𝑎𝑦, 𝐷𝑎𝑦𝑂𝑓𝑌𝑒𝑎𝑟,𝐷𝑎𝑡𝑒} features from 𝐼𝑛𝑣𝑜𝑖𝑐𝑒𝐷𝑎𝑡𝑒 feature :

```python
c["Year"] = c['InvoiceDate'].dt.year
c["Quarter"] = c['InvoiceDate'].dt.quarter
c["Month"] = c['InvoiceDate'].dt.month
c["Week"] = c['InvoiceDate'].dt.isocalendar().week
c["Weekday"] = c['InvoiceDate'].dt.dayofweek
c["Day"] = c['InvoiceDate'].dt.day
c["Dayofyear"] = c['InvoiceDate'].dt.dayofyear
c["Date"] = c['InvoiceDate'].dt.date
c.head()
```

### 7) Aggregate Revenue across {𝐈𝒏𝒗𝒐𝒊𝒄𝒆𝑫𝒂𝒕𝒆,𝑺𝒕𝒐𝒄𝒌𝑪𝒐𝒅𝒆,Date extracted features} :

```python
grouped_features1= c.groupby(['InvoiceDate','StockCode','Year','Quarter','Month','Week','Weekday','Day','Dayofyear','Date']).Revenue.sum().reset_index()
grouped_features1
```

### 8) Aggregate Quantity across {𝐈𝒏𝒗𝒐𝒊𝒄𝒆𝑫𝒂𝒕𝒆,𝑺𝒕𝒐𝒄𝒌𝑪𝒐𝒅𝒆,Date extracted features} :

```python
grouped_features2= c.groupby(['InvoiceDate','StockCode','Year','Quarter','Month','Week','Weekday','Day','Dayofyear','Date']).Quantity.sum().reset_index()
grouped_features2
```
